```python
# Bibliotecas
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

## Distribui√ß√£o Uniforme

+ `stats.uniform.pdf(ùë•, loc = ùëé, scale = ùëè ‚àí ùëé)`
+ `stats.uniform.cdf(ùë•, loc = ùëé, scale = ùëè ‚àí ùëé)`

- $ a = $ menor valor
- $ b = $ maior valor
- $ x_u = $ valor entre $a$ e $b$
- $x_u \in \square a, b \rbrace$ 

```python

plt.figure(figsize=(14, 4.5))
xu = np.linspace(0, 10, 1000)

fxu = stats.uniform.pdf(xu, loc=0, scale=10)
plt.subplot(121)
plt.plot(xu, fxu, 'm.-')
plt.title('stats.uniform.pdf', fontsize=16, color='black')

fxu = stats.uniform.cdf(xu, loc=0, scale=10)
plt.subplot(122)
plt.plot(xu, fxu, 'm.-')
plt.title('stats.uniform.cdf', fontsize=16, color='black')

plt.show()
```

## Distribui√ß√£o Exponencial

- `stats.expon.pdf(0.5, loc = 0, scale = mu)`
- `stats.expon.cdf(0.5, loc = 0, scale = mu)`
- `stats.expon.ppf(0.5, loc = 0, scale = mu)`

- $\mu =$ tempo m√©dio para evento ocorer
- $\lambda = \frac{1}{\mu}$ = frequ√™ncia
- $x_e \in \lbrace 0, +\infty \rbrace$ 


```python

plt.figure(figsize=(14, 4.5))
xe = np.linspace(0, 200, 1000)

fxe = stats.expon.pdf(xe, 0, 20)
plt.subplot(121)
plt.plot(xe, fxe, 'c.-')
plt.title('stats.expon.pdf', fontsize=16, color='black')

fxe = stats.expon.cdf(xe, 0, 20)
plt.subplot(122)
plt.plot(xe, fxe, 'c.-')
plt.title('stats.expon.cdf', fontsize=16, color='black')

plt.show()
```

## Distribui√ß√£o Normal

```python

plt.figure(figsize=(14, 4.5))
xn = np.linspace(0, 10, 1000)

fxn = stats.norm.pdf(xn, 5, 1)
plt.subplot(121)
plt.plot(xn, fxn, 'k.-')
plt.title('stats.norm.pdf', fontsize=16, color='black')

fxn = stats.norm.cdf(xn, 5, 1)
plt.subplot(122)
plt.plot(xn, fxn, 'k.-')
plt.title('stats.norm.cdf', fontsize=16, color='black')

plt.show()
```
