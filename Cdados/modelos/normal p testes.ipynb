```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt 
```

```python
xn1 = np.linspace(900, 2000, 1000)
fxn1 = stats.norm.pdf(xn1, 1400, 45)
fxn2 = stats.norm.pdf(xn1, 1400, 90)
fxn3 = stats.norm.pdf(xn1, 1400, 135)

plt.plot(xn1, fxn1)
plt.plot(xn1, fxn2)
plt.plot(xn1, fxn3)
plt.show()
```

```python
figure = plt.figure(figsize=(10,6.7))
L = np.arange(45, 135+1, 45)
xn1 = np.linspace(800, 2000, 1000)

for i in L:
    fxn1 = stats.norm.pdf(xn1, 1400, i)
    plt.plot(xn1, fxn1)
plt.show()

figure = plt.figure(figsize=(10,6.7))
L = np.arange(45, 135+1, 45)
for i in L:
    xn1 = np.linspace(800, 2000, 1000)
    xn1 = (xn1-1400)/i
    fxn1 = stats.norm.pdf(xn1)
    plt.plot(xn1, fxn1)

plt.show()
```
